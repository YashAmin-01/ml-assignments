# Exception Handling Assignment

```python
def divide(a,b):
    try:
        c = a/b
        return c
    except Exception as e:
        return "Number cannot be divided by zero"
```

```python
divide(5,3)
```

```python
divide(5,0)
```

```python
subjects = ['Americans', 'Indians']
verbs = ['play', 'watch']
objects = ['baseball', 'cricket']
```

```python
for i in subjects:
    for j in verbs:
        for k in objects:
            print(i + " " + j + " " + k)
```
