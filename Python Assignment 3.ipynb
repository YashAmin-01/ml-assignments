```python
list1 = [2,5,6,8,7,4,1,6,9,8,3,5]
```

```python
functools.reduce(lambda a,b: a+b, list1)
```

```python
def myreduce(func,iterable):
    a = iterable[0]
    for i in range(1,len(iterable)):
        b = iterable[i]
        a = func(a,b)
    return a
def add(a,b):
    return a+b

print(myreduce(add,list1))
```

```python
def is_string(s):
    if type(s) == str:
        return True
    else:
        return False
```

```python
list2 = [1,2,'yash',4,5,'amin',9]
```

```python
print(list(filter(is_string, list2)))
```

```python
def myfilter(func,iterable):
    list_out = []
    for i in iterable:
        if func(i):
            list_out.append(i)
    print(list_out)

myfilter(is_string,list2)
```
