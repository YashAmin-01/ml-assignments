```python
for i in range(5):
    for j in range(i):
        print("*", end="")
    print()

for i in range(5):
    for j in range(5-i):
        print("*", end="")
    print()
```

```python
word = input("Enter the word: ")

reverse = ""

for i in word:
    reverse = i + reverse
print(reverse)
```
