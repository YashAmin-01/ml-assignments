```python
class Sides():
    def __init__(self):
        a = 0
        b = 0
        c = 0
        
    def sides_input(self):
        a = int(input("enter value of side a: "))
        b = int(input("enter value of side b: "))
        c = int(input("enter value of side c: "))
        return a,b,c
```

```python
class Area(Sides):
    def __init__(self):
        s = 0
        area = 0
        
    def calc_s(self):
        s = (a + b + c) / 2
        return s
        
    def calc_area(self, s):
        area = (s*(s-a)*(s-b)*(s-c)) ** 0.5
        return area
```

```python
sides = Sides()
a,b,c = sides.sides_input()
print(a, b, c)
```

```python
area = Area()
s = area.calc_s()
triangle_area = area.calc_area(s)
```

```python
print(triangle_area)
```

```python
def filter_long_words(word_list, n):
    new_word_list = []
    for i in word_list:
        if len(i) > n:
            new_word_list.append(i)
    return new_word_list
```

```python
word_list = ['lion', 'tiger', 'elephant', 'cat', 'dog']

print(filter_long_words(word_list, 4))
```

```python
def mapWordToLength(word_list):
    length_list = []
    
    for i in word_list:
        length_list.append(len(i))
    
    return length_list
```

```python
print(mapWordToLength(word_list))
```

```python
def is_vowel(char):
    vowels = ['a', 'e', 'i', 'o', 'u']
    
    if char in vowels:
        return True
    else:
        return False
```

```python
print(is_vowel('a'))
print(is_vowel('c'))
print(is_vowel('d'))
print(is_vowel('e'))
```
